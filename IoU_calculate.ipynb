```python
import numpy as np
import cv2
import os

iou_sum = 0.0
imageID = 0
for root, dirs, files in os.walk("C:/Users/xiesw/Desktop/Scribble_deepglobe/true_mask_valid"):
    for file in files:
        imageID = imageID + 1
        print(imageID)
        mask_true = cv2.imread("true_mask_valid/" + file)
        mask_true = mask_true[:,:,0] / 255
        mask_true = mask_true.astype(int)
        mask_pre = cv2.imread("post_processed_predictions/" + file[:-4] + "_post_processed.png")
        #mask_pre = cv2.imread("predictions/" + file)
        mask_pre = mask_pre[:,:,0] / 255
        mask_pre = mask_pre.astype(int)
        intersection = (mask_true & mask_pre).sum((0,1))
        union = (mask_true | mask_pre).sum((0,1))
        iou = (intersection + 1e-6) / (union + 1e-6)
        print(iou)
        iou_sum = iou_sum + iou
iou_mean = iou_sum / imageID
print(iou_mean)
```
